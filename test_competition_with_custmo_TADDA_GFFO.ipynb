```python
import pandas as pd
import numpy as np

```

```python
def calculate_mean_restorename(dataset,stat_name):
    dataset_mean=dataset.copy()
    dataset_mean = dataset_mean.mean(axis="columns")
    dataset_mean = pd.DataFrame(dataset_mean.values,columns=[stat_name])
    dataset_mean["model"] = dataset["Unnamed: 0"]
    return dataset_mean


```

```python
t1_cm_mse=pd.read_csv("output/tables/t1_cm_mse.csv")
t1_cm_mse_nonzero=pd.read_csv("output/tables/t1_cm_mse_nonzero.csv")
t1_cm_tadda_1=pd.read_csv("output/tables/t1_cm_tadda_1.csv")
t1_cm_tadda_1_nonzero=pd.read_csv("output/tables/t1_cm_tadda_1_nonzero.csv")
t1_cm_tadda_2=pd.read_csv("output/tables/t1_cm_tadda_2.csv")
t1_cm_tadda_2_nonzero=pd.read_csv("output/tables/t1_cm_tadda_2_nonzero.csv")
t1_cm_tadda_gffo=pd.read_csv("output/tables/t1_cm_tadda_gffo.csv")
```

```python
t1_cm_mse_mean=calculate_mean_restorename(t1_cm_mse,"t1_cm_mse")
t1_cm_mse_nonzero_mean=calculate_mean_restorename(t1_cm_mse_nonzero,"t1_cm_mse_nonzero")
t1_cm_tadda_1_mean=calculate_mean_restorename(t1_cm_tadda_1,"t1_cm_tadda_1")
t1_cm_tadda_1_nonzero_mean=calculate_mean_restorename(t1_cm_tadda_1_nonzero,"t1_cm_tadda_1_nonzero")
t1_cm_tadda_2_mean=calculate_mean_restorename(t1_cm_tadda_2,"t1_cm_tadda_2")
t1_cm_tadda_2_nonzero_mean=calculate_mean_restorename(t1_cm_tadda_2_nonzero,"t1_cm_tadda_2_nonzero")
t1_cm_tadda_gffo_mean=calculate_mean_restorename(t1_cm_tadda_gffo,"t1_cm_tadda_gffo")

out_t1 = t1_cm_mse_mean.merge(t1_cm_mse_nonzero_mean, on='model', how='left')
out_t1 = out_t1.merge(t1_cm_tadda_1_mean, on='model', how='left')
out_t1 = out_t1.merge(t1_cm_tadda_1_nonzero_mean, on='model', how='left')
out_t1 = out_t1.merge(t1_cm_tadda_2_mean, on='model', how='left')
out_t1 = out_t1.merge(t1_cm_tadda_2_nonzero_mean, on='model', how='left')
out_t1 = out_t1.merge(t1_cm_tadda_gffo_mean, on='model', how='left')
out_t1=out_t1.sort_values(by=['t1_cm_tadda_gffo'])
out_t1 = out_t1[["model","t1_cm_mse","t1_cm_mse_nonzero","t1_cm_tadda_1","t1_cm_tadda_1_nonzero","t1_cm_tadda_2","t1_cm_tadda_2_nonzero","t1_cm_tadda_gffo"]]
```

```python
out_t1
```

```python
out_t1.to_csv("task1.csv")
```
